```python
df['Year_Parsed'] = df['Year'].str[:4].astype(int)

# Filter for General category
general_df = df[(df['Type'] == 'General') & (df['Category'] == 'General')]

# Plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=general_df, x='Year_Parsed', y='Index', marker='o', label='Index')
sns.lineplot(data=general_df, x='Year_Parsed', y='Growth Rate (%)', marker='s', label='Growth Rate (%)')
plt.title('General IIP Index and Growth Rate Over Time')
plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### The plot above shows the General Index of Industrial Production (IIP) and its Growth Rate (%) over time:

* The Index generally increases over the years, indicating industrial expansion.

* The Growth Rate (%) shows more fluctuation, with noticeable ups and downs reflecting economic variability

```python
# Filter Sectoral data
sectoral_df = df[df['Type'] == 'Sectoral']

# Plot
plt.figure(figsize=(12, 6))
for sector in sectoral_df['Category'].unique():
    sector_data = sectoral_df[(sectoral_df['Category'] == sector) & (sectoral_df['SubCategory'] == '*')]
    sns.lineplot(data=sector_data, x='Year_Parsed', y='Index', marker='o', label=sector)

plt.title('Sectoral IIP Index Over Time')
plt.xlabel('Year')
plt.ylabel('Index')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### This plot shows the Sectoral Index trends over time for the main sectors:

* Manufacturing has the highest and most variable index, indicating it's the largest industrial contributor.

* Electricity has a steady upward trend, reflecting consistent infrastructure development.

* Mining shows slower, more fluctuating growth, likely tied to global commodity demand and policy.

```python
data=df[df["Category"]=="Electricity"]

px=sns.barplot(data=data.sort_values(by=["Year"]),x="Year",y="Growth Rate (%)",hue="Category")

px.set_title("Year wise electricity growth rate chart")

plt.xticks(rotation=90)
```

#### Above chart shows year wise electricity consumption growth in last 12 years.
* Clearly,we see dip in electricity consumption during year 2020-21.
* This decline in consumption can be attributed to covid.

```python
data_manufacturing=df[df["Category"]=="Manufacturing"]

px=sns.lineplot(data=data_manufacturing.sort_values(by=["Year"]),x="Year_Parsed",y="Growth Rate (%)",hue="Category")

px.set_title("Year wise manufacturing growth rate chart")

plt.xticks(rotation=90)

plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')
```

#### The above chart shows the annual growth rate of manufacturing sector in last 12 years
* We see huge dip in growth in year 2020,again which can be attributed to covid pandemic.
